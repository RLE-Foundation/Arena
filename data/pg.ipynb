````python
import pandas as pd
import numpy as np
import json

data = pd.read_json('procgen/procgen_data.json')
# reset the index and name it to 'env'
data = data.reset_index()
data.rename(columns={'index': 'env'}, inplace=True)
data
````

````python
env_data = data[data['env'] == 'bigfish']
env_data
````

````python
all_data = {
    'Easy': {},
    'Hard': {}
}

for env_name in data['env'].unique():
    env_data = data[data['env'] == env_name]
    json_data = []
    data_piece = {
        'Team': 'Google Brain',
        'Name': None,
        'Version': 'Hard',
        'Training Steps': '25M',
        'Parameters': 1000000,
        'Score': None,
        'Mean': None,
        'Code': 'https://github.com/google-research/rliable',
        'Date': '2022-01-01'
    }

    for algo in env_data.columns[1:]:
        # print(algo)
        data_piece['Name'] = algo
        data_piece['Score'] = env_data[algo].values[0]
        data_piece['Mean'] = np.mean(data_piece['Score']).round(3)
        json_data.append(data_piece.copy())
    
    all_data['Hard'][env_name] = json_data

    # save json_data to bigfish.json
with open(f'../docs/datasets/procgen.json', 'w') as f:
    json.dump(all_data, f)
````
